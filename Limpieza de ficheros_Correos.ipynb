### 1. Carga de datos

Existen ditintos tipos de ficheros en función del tipo de movimiento:
- **Llegada**: dan información relacionada con los vuelos que aterrizan/llegan a los aeropuertos
- **Salida**: dan información relacionada con los vuelos que despegan/salen a los aeropuertos
  
A su vez, cada uno de ellos está divido en:

- **Nacional**: el origen y el destino del vuelo es español
- **Total**: se incluyen todos los origenes y destinos posibles (nacional, internacional, Unión Europea No Schengen, Unión Europea Schengen, Schengen No Unión Europea o No Schengen No Unión Europea).

Los datos abarcan desde 2004 hasta junio de 2024, almacenados en ficheros de dos años, por lo que se dispone de 11 ficheros para cada uno de los tipos, 11 de llegada nacionales, 11 llegada totales, 11 de salida nacionales y 11 de salida totales.

Todas las columnas son números enteros y no hay ningún nulo.

Como va a haber muchos ficheros que tendrán las columnas con nombres similares se renombran las columnas.

```python
# Se renombran las columnas mediante una función
def rename_column_total_llegada(col):
    if col != 'Aeropuerto Base':  # No se renombra la primera columna 'Aeropuerto Base'
        # Se reemplaza "de" por un espacio vacío, los espacios vacios por un guion bajo y, se usan las tres primeras letras del mes
        parts = col.split()
        month_abbrev = parts[0][:3]  # Se toman las primeras tres letras del mes
        year = parts[-1]  # Año
        return f"{month_abbrev}_{year}_C_T_Lleg" # Se añade al final un _T_Lleg, para identificarlas como el fichero de datos totales de llegada
    else:
        return col

# Se aplica la función a todas las columnas
F_Corr_T_llegada_0405.columns = [rename_column_total_llegada(col) for col in F_Corr_T_llegada_0405.columns]

# Se muestran las primeras filas para ver si está correcto el formato del fichero de totales por llegada
F_Corr_T_llegada_0405.head()
```

### 3. Carga de toda la carpeta de Correos_Total_Llegada

Como hay ficheros separados en 2 años desde 2004, es decir, 11 ficheros, se va a tratar cargarlos todos desde la carpeta, para no tener que repetir el proceso tantas veces como ficheros haya.

```python
# Se guarda la ruta de la carpeta que contiene los ficheros de correos de llegada totales
Ruta_C_Llegada_Total = './Correos/C_Llegada_Total/'

# Se obtiene una lista de todos los ficheros que hay en la carpeta
Ficheros_C_Llegada_Total = [f for f in os.listdir(Ruta_C_Llegada_Total) if f.endswith('.csv')]

# Se quiere guardar cada dataframe por separado por lo que se crea un diccionario
Dicc_C_Llegada_Total = {}

# Al igual que se hizo con el archivo de 2004_2005 se renombran las columnas de los ficheros, se usa la misma función
def rename_column_total_llegada_C(col):
    if col != 'Aeropuerto Base': # No se renombra la primera columna 'Aeropuerto Base'
        # Se reemplaza "de" por un espacio vacío, los espacios vacios por un guion bajo y, se usan las tres primeras letras del mes
        parts = col.split()
        month_abbrev = parts[0][:3] # Se toman las primeras tres letras del mes
        year = parts[-1]
        return f"{month_abbrev}_{year}_C_T_Lleg" # Se añade al final un _T_Lleg, para identificarlas como el fichero de datos totales de llegada
    else:
        return col

# Se carga cada archivo en un dataframe separado
for Ficheros_C_LL_T in Ficheros_C_Llegada_Total:
    # Nombre del dataframe: nombre del archivo sin la extensión
    Nombre_Df_C_Llegada_Total = os.path.splitext(Ficheros_C_LL_T)[0]
    
    try: # Se añade una excepción por si no se lee bien un fichero
        Df_C_Llegada_Total_f = pd.read_csv(os.path.join(Ruta_C_Llegada_Total, Ficheros_C_LL_T), skiprows=1, encoding='utf-16', delimiter=",", thousands=".")
    except Exception as e: 
        print(f"Error al procesar el archivo {Ficheros_C_LL_T}: {e}")
        continue
       
    # Se renombran las columnas según la función "rename_column_total_llegada_M"
    Df_C_Llegada_Total_f.columns = [rename_column_total_llegada_C(col) for col in Df_C_Llegada_Total_f.columns]

    # Se renombra la columna "Aeropuerto Base" a "Aeropuerto_Base"
    Df_C_Llegada_Total_f.rename(columns={"Aeropuerto Base": "Aeropuerto_Base"}, inplace=True)
    
    # Se almacena cada dataframe (fichero) en el diccionario "Dicc_C_Llegada_Total"
    Dicc_C_Llegada_Total[Nombre_Df_C_Llegada_Total] = Df_C_Llegada_Total_f

# Se verifica que se han cargado todos los ficheros y con los nombres de columnas esperados
for key, df in Dicc_C_Llegada_Total.items():
    print(f"{key} cargado con {df.shape[0]} filas y {df.shape[1]} columnas.")
```

Se inspeccionan los ficheros y todos parecen tener el mismo número de columnas, la diferencia de filas ya se especifica en el documento y proviene de la disponibilidad de datos de AEna para esos aeropuertos o debido a su fecha de comienzo a operar. Cada uno de los ficheros tiene 2 años, 24 columnas + la columna de Aeropuerto_Base, salvo para el año 2024 que unicamente tiene datos hasta junio.

**Crear un único dataframe con los correos totales de llegada**

Como en principio no presentan diferencias, se va a generar un único dataframe por los movimientos de llegada de pasajeros totales, el proceso es hacer un merge por la columna de "Aeropuerto Base", para tener la información de cada uno de los aeropuertos por año.

Se verifica que todos los ficheros tienen la columna de "Aeropuerto_Base" ya que es la que se va a utilizar para unirlos.

```python
todos_tienen_aeropuerto_base = True

for fichero, df in Dicc_C_Llegada_Total.items():
    if 'Aeropuerto_Base' not in df.columns:
        print(f"El fichero '{fichero}' no tiene la columna 'Aeropuerto_Base'.")
        print(f"Las columnas disponibles en este fichero son: {df.columns.tolist()}")
        todos_tienen_aeropuerto_base = False

# Mensaje si todos los DataFrames tienen la columna "Aeropuerto_Base"
if todos_tienen_aeropuerto_base:
    print("Todos los ficheros tienen la columna 'Aeropuerto_Base'.")
```

```python
# Para poder unir los dataframes primero es necesario convertir todo lo que se ha almacenado en el diccionario (nombre de los ficheros) hay que crear una lsita
Lista_C_Llegada_Total = list(Dicc_C_Llegada_Total.values())

# Se unen los dataframes por la columna de "Aeropuerto Base"
Df_C_Llegada_Total_SM = reduce(lambda left, right: pd.merge(left, right, on='Aeropuerto_Base', how='outer'), Lista_C_Llegada_Total)

# Se muestran las primeras filas
Df_C_Llegada_Total_SM.head()
```

```python
# Convertir todas las columnas de tipo flotante a enteros
Df_C_Llegada_Total_SM = Df_C_Llegada_Total_SM.apply(lambda x: x.fillna(0).astype(int) if x.dtype == 'float64' else x)

# Mostrar las primeras filas del DataFrame resultante
Df_C_Llegada_Total_SM.head()
```

Tras haber hecho una inspección manual de los datos, y que se incluye en el documento, hay nulos en los valores de Aeropuerto Intl. Región Murcia, Ageciras/helipuerto, Burgos, Huesca-Pirineos, Madrid-Torrejón, Murcia San Javier, Son Bonet y Sabadell.

Se opta por unir los datos de los aeropuertos de Aeropuerto Intl. Región Murcia y Murcia San Javier ya que el desdoblamiento se produce por como cambió de categoría el aeropuerto.

```python
# Se filtran las filas correspondientes a los aeropuertos de Murcia
Murcia_aeropuertos = Df_C_Llegada_Total_SM[
    Df_C_Llegada_Total_SM['Aeropuerto_Base'].isin(['AEROPUERTO INTL. REGIÓN MURCIA', 'MURCIA-SAN JAVIER'])
]

# Se suman los datos de las filas de esos aeropuertos, excluyendo la columna 'Aeropuerto_Base'
Murcia_filas_combinadas = Murcia_aeropuertos.drop(columns='Aeropuerto_Base').sum()

# Se crea una nueva fila con el nombre "MURCIA COMBINADO"
Murcia_filas_combinadas['Aeropuerto_Base'] = 'MURCIA COMBINADO'
Murcia_filas_combinadas = pd.DataFrame([Murcia_filas_combinadas])

# Se eliminan las filas originales de los aeropuertos
Df_C_Llegada_Total_SM = Df_C_Llegada_Total_SM[
    ~Df_C_Llegada_Total_SM['Aeropuerto_Base'].isin(['AEROPUERTO INTL. REGIÓN MURCIA', 'MURCIA-SAN JAVIER'])
]

# Se añade la nueva fila al dataframe
Df_C_Llegada_Total = pd.concat([Df_C_Llegada_Total_SM, Murcia_filas_combinadas], ignore_index=True)

# Se comprueba que se haya realizado correctamente
Df_C_Llegada_Total.loc[Df_C_Llegada_Total['Aeropuerto_Base'] == 'MURCIA COMBINADO']
```

Por último, se van a ordenar las columnas cronológicamente

```python
# Se define la columna que no se va a tener en cuenta en la ordenación
columna_base = 'Aeropuerto_Base'

# Se extraen las columnas temporales sin contar con la columna base: "Aeropuerto_base"
columnas_temporales_C_T = [col for col in Df_C_Llegada_Total.columns if col != columna_base]

# Diccionario para convertir abreviaciones de meses a números
Meses_A_Numeros = {
    'Ene': '01', 'Feb': '02', 'Mar': '03', 'Abr': '04',
    'May': '05', 'Jun': '06', 'Jul': '07', 'Ago': '08',
    'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dic': '12'
}

# Se define una función para convertir las columnas en temporales
def convertir_fecha(col_name):
    Mes_Numero, Año = col_name.split('_')[:2] # Se separa el nombre de las columnas por el guion bajo
    Mes = Meses_A_Numeros[Mes_Numero] # Se cambia el texto por el número del mes
    return pd.to_datetime(f'{Año}-{Mes}', format='%Y-%m')

# Se ordenan las columnas cronologicamente
Columnas_Cronologicas_C_T = sorted(columnas_temporales_C_T, key=convertir_fecha)

# Se crea el nuevo orden de columnas
Columnas_Ordenadas_Cro_C_T = [columna_base] + Columnas_Cronologicas_C_T

# Se reordena el dataframe para que vaya del menor mes y año hasta el mayor
Df_C_Llegada_Total = Df_C_Llegada_Total[Columnas_Ordenadas_Cro_C_T]

# Se verifica que se ha ordenado correctamente
Df_C_Llegada_Total.head()
```

**Se modifican los nombre de las columnas y se pivota la tabla**

```python
# Se definen los tipos de columnas que va a tener la tabla despoués de pivotarla
Correos_merge_pivotar = pd.melt(Correos_merge, 
                                 id_vars=['Aeropuerto_Base'], 
                                 var_name='Fecha', 
                                 value_name='Valor')

# Se verifica que se ha pivotado
print(Correos_merge_pivotar.head())
```

Mediante el sufijo de la columna fecha se genera una nueva columna "Tipo" para diferenciar que dato es el que se está mostrando, si los datos totales de llegada, totales de salida, nacionales de llegada o nacionales de salida.

```python
# Se crea una nueva columna para no perder el tipo de operación (total llegada, nacional llegada, total salida y nacional salida)
Correos_merge_pivotar['Tipo'] = Correos_merge_pivotar['Fecha'].apply(lambda x: 'C_Total_Llegada' if '_T_Lleg' in x else 
                                                                             'C_Nacional_Llegada' if '_N_Lleg' in x else 
                                                                             'C_Total_Salida' if '_T_Sal' in x else 
                                                                             'C_Nacional_Salida')

# Se comprueba que se mapea correctamente las columnas "fecha y tipo" en función del sufijo
print(Correos_merge_pivotar[['Fecha', 'Tipo']].head())
```

```python
# Se elimina de la fecha los sufijos, ya que para luego mapear el campo de fecha "molesta"
Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].str.replace('_C_T_Lleg', '', regex=False)
Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].str.replace('_C_N_Lleg', '', regex=False)
Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].str.replace('_C_T_Sal', '', regex=False)
Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].str.replace('_C_N_Sal', '', regex=False)

# Se comprueba que se han eliminado correctamente el sufijo y se muestran todos los valores únicos
print(Correos_merge_pivotar['Fecha'].unique())
```

```python
# Se mapean los meses en español a inglés, paso necesario para luego poder a convertir la columna a formato datatime
meses_espanol = {'Ene': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Abr': 'Apr', 'May': 'May', 'Jun': 'Jun',
                 'Jul': 'Jul', 'Ago': 'Aug', 'Sep': 'Sep', 'Oct': 'Oct', 'Nov': 'Nov', 'Dic': 'Dec'}

# Se remplazan los nombres de español a inglés
for mes_es, mes_en in meses_espanol.items():
    Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].str.replace(mes_es, mes_en)

# Se comprueba que se realiza correctamente el cambio de la fecha de español a inglés
print(Correos_merge_pivotar.head())
```

```python
# Se comvierte la fecha a formato datetime
Correos_merge_pivotar['Fecha'] = pd.to_datetime(Correos_merge_pivotar['Fecha'], format='%b_%Y')

# Se comprueba que se realiza correctamente
print(Correos_merge_pivotar.head())
```

```python
# Se quita de la columna fecha el día ya que no aporta valor
Correos_merge_pivotar['Fecha'] = Correos_merge_pivotar['Fecha'].dt.to_period('M')

# Se comprueba que se realiza correctamente
print(Correos_merge_pivotar.head())
```
